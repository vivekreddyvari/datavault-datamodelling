```python
from pyspark.sql import SparkSession
from pyspark.sql import types as T
from pyspark.sql import functions as F
import logging
from dataclasses import dataclass

class ColorFormatter(logging.Formatter):
    COLORS = {
        logging.ERROR: '\033[91m',   # Red
        logging.INFO: '\033[93m',    # Yellow
    }
    ICONS = {
        logging.ERROR: '\u26A0\uFE0F ERROR:',  # ⚠️ ERROR:
        logging.INFO: '\u2139\uFE0F INFO:',    # ℹ️ INFO:
    }
    RESET = '\033[0m'

    def format(self, record):
        color = self.COLORS.get(record.levelno, '')
        icon = self.ICONS.get(record.levelno, '')
        msg = super().format(record)
        return f"{color}{icon} {msg}{self.RESET}"

@dataclass
class HELPER:
    """
    A class containing helper functions for Spark operations, enhanced with color and icon output.
    """
    RED = '\033[91m'
    YELLOW = '\033[93m'
    RESET = '\033[0m'

    @staticmethod
    def print_error(msg: str):
        print(f"{HELPER.RED}\u26A0\uFE0F ERROR: {msg}{HELPER.RESET}")

    @staticmethod
    def print_info(msg: str):
        print(f"{HELPER.YELLOW}\u2139\uFE0F INFO: {msg}{HELPER.RESET}")

    def get_spark_session(self, name: str):
        return SparkSession.builder.appName(name).getOrCreate()

    def get_logger(self, name: str):
        logger = logging.getLogger(name)
        logger.setLevel(logging.DEBUG)
        ch = logging.StreamHandler()
        ch.setLevel(logging.DEBUG)
        formatter = ColorFormatter('%(message)s')
        ch.setFormatter(formatter)
        if not logger.hasHandlers():
            logger.addHandler(ch)
        return logger
```
